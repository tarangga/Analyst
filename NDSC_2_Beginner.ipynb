```
import pandas as pd
rules = pd.read_csv('/content/rules.csv')
rules.info()
for r in rules.rule:
  print(r)
```

```
order = pd.read_csv('/content/association_order.csv')
order.info()
order.head()
```

```
def confidence(rule):
  r = rule.split('>')
  arr = r[0].split('&') + r[1].split('&')
  first = order[order.itemid == int(arr[0])]['orderid']
  second = order[order.itemid == int(arr[1])]['orderid']
  merge = pd.merge(first, second)
  if len(arr) == 2:
    num2 = merge.shape[0]
  else:
    third = order[order.itemid == int(arr[2])]['orderid']
    num2 = pd.merge(merge, third).shape[0]

  return num2/first.shape[0]
```

```
result = rules.rule.apply(confidence)
```

```
import numpy as np

rules['confidence'] = np.floor(result*1000).astype(int)
rules.to_csv('rule_.csv', index=False)
```
